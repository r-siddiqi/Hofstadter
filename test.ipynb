```python
from lattice_types.honeycomb import Honeycomb_Hamiltonian
from lattice_types.square import Square_Hamiltonian
from lattice_types.plotting import Plot_Lattice
```

The following is an example of using the imported square and honeycomb lattice classes to generate some plots using the imported plotting functions

```python
#Initialize the Honeycomb
Honeycomb = Honeycomb_Hamiltonian(10, 10, 10, 10, 10)

#Generate outputs of the honeycomb pass through to plot
honeycomb_outputs = Honeycomb.prepare_outputs()

#Initializing Honeycomb plotting
honeycomb_plot = Plot_Lattice(honeycomb_outputs)

#Plot the Eigenvalue spectrum of the Honeycomb
honeycomb_plot.plot_evals()
```
