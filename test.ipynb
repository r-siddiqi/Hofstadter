```python
from lattice_types.honeycomb import Honeycomb_Hamiltonian
from lattice_types.square import Square_Hamiltonian
from lattice_types.kagome import Kagome_Hamiltonian
from lattice_types.plotting import Plotting_Functions
```

The following is an example of using the imported square and honeycomb lattice classes to generate some plots using the imported plotting functions

```python
#Define Simulation Parameters

length_h = 10   # Lattice dimension
t_h = 1         # Hopping parameter
W_h = 0.5       # Disorder strength
phi_h = 0.5     # Magnetic flux per plaquette
q_h = 50        # Maximum denominator for phi values


#Initialize the Honeycomb

Honeycomb = Honeycomb_Hamiltonian(length_h, t_h, W_h, phi_h, q_h, save = True)

#Generate outputs of the honeycomb pass through to plot
honeycomb_outputs = Honeycomb.prepare_outputs()

#Initializing Honeycomb plotting
honeycomb_plot = Plotting_Functions(honeycomb_outputs, save = True)

#Plot the Eigenvalue spectrum of the Honeycomb
honeycomb_plot.plot_evals(save = True)

honeycomb_plot.plot_density_of_states(save = True)

Honeycomb.plot_hofstadter_butterfly(save = True)
```

```python
#Initialize the Square
Square = Square_Hamiltonian(10, 1, 0.5, 0.0, 50, save = True)

#Generate outputs of the square pass through to plot
square_outputs = Square.prepare_outputs()

#Initializing Honeycomb plotting
square_plot = Plotting_Functions(square_outputs, save = True)

#Plot the Eigenvalue spectrum of the Square
square_plot.plot_evals(save = True)

Square.plot_hofstadter_butterfly(save = True)


```

```python
#Initialize the Square
Kagome = Kagome_Hamiltonian(t=1, W= 0, phi = 0.05, q = 50, save = True)

#Generate outputs of the square pass through to plot
kagome_outputs = Kagome.prepare_outputs()

#Initializing Honeycomb plotting
kagome_plot = Plotting_Functions(kagome_outputs, save = True)

#Plot the Eigenvalue spectrum of the Square
kagome_plot.plot_evals(save = True)

Kagome.plot_hofstadter_butterfly(save = True)
```
